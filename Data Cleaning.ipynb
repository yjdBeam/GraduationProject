# 4. 数据筛选

### 4.1 筛选列

```python
df_col = df[['accepted_rank','genus','phylum','class','order','family','environment','lithification1']]
```

```python
df_col
```

### 4.2 Taxonomic filtering

##### 4.2.1 accepted rank

```python
df_ar = df_col[df_col['accepted_rank'].isin(['genus','species','subgenus'])]
```

```python
df_ar
```

```python
df_genus = df_ar[df_ar['genus'] != '']
```

```python
df_genus
```

##### 4.2.2 Phylum

```python
df_phylum = df_genus[df_genus['phylum'].isin(["Agmata","Annelida","Bilateralomorpha","Brachiopoda","Bryozoa","Calcispongea",
                                               "Chaetognatha","Cnidaria","Ctenophora","Echinodermata","Entoprocta","Foraminifera",
                                               "Hemichordata","Hyolitha","Mollusca","Nematoda","Nematomorpha","Nemertina",
                                               "Onychophora","Petalonamae","Phoronida","Platyhelminthes","Porifera","Problematica",
                                               "Rhizopodea","Rotifera","Sarcomastigophora","Sipuncula","Uncertain","Vetulicolia"])]
```

```python
df_phylum
```

##### 4.2.3 Class

```python
df_class = df_genus[df_genus['class'].isin(["Acanthodii","Actinopteri","Actinopterygii","Agnatha","Cephalaspidomorphi",
                                              "Chondrichthyes","Cladistia","Coelacanthimorpha","Conodonta","Galeaspida","Myxini",
                                              "Osteichthyes","Petromyzontida","Plagiostomi","Pteraspidomorphi","Artiopoda",
                                              "Branchiopoda","Cephalocarida","Copepoda","Malacostraca","Maxillopoda","Megacheira",
                                              "Merostomoidea","Ostracoda","Paratrilobita","Pycnogonida","Remipedia", "Thylacocephala",
                                              "Trilobita","Xiphosura"])]
```

```python
df_class
```

##### 4.2.4 Mammals and Carnivores

```python
df_of = df_genus[df_genus['order'].isin(["Cetacea", "Sirenia"]) | 
                 df_genus['family'].isin(["Otariidae", "Phocidae", "Desmatophocidae"])]
```

```python
df_of
```

##### 4.2.5 Reptiles

```python
df_order = df_genus[df_genus['order'].isin(["Eosauropterygia","Hupehsuchia","Ichthyosauria",
                                            "Placodontia","Sauropterygia","Thalattosauria"])]
```

```python
df_order
```

##### 4.2.6 Sea turtles

```python
df_family = df_genus[df_genus['family'].isin(["Cheloniidae","Protostegidae","Dermochelyidae",
                                              "Dermochelyoidae","Toxochelyidae","Pancheloniidae"])]
```

```python
df_family
```

```python
df_taxonomic = pd.concat([df_phylum,df_class,df_of,df_order,df_family])
```

```python
df_taxonomic
```

### 4.3 Sedimentary environment filtering

##### 4.3.1 Omit environment

```python
df_environment = df_taxonomic[~df_taxonomic['environment'].isin(['\"floodplain\"', "alluvial fan", "cave", '\"channel\"', 
                                                                 "channel lag" , "coarse channel fill", "crater lake", "crevasse splay", 
                                                                 "dry floodplain", "delta plain", "dune", "eolian indet.", 
                                                                 "fine channel fill", "fissure fill", "fluvial indet.", 
                                                                 "fluvial-lacustrine indet.", "fluvial-deltaic indet.", "glacial", 
                                                                 "interdune", "karst indet.", "lacustrine - large", "lacustrine - small",
                                                                 "lacustrine delta front", "lacustrine delta plain", 
                                                                 "lacustrine deltaic indet.", "lacustrine indet.",  
                                                                 "lacustrine interdistributary bay", "lacustrine prodelta", "levee", 
                                                                 "loess", "mire/swamp", "pond", "sinkhole", "spring", "tar", 
                                                                 "terrestrial indet.", "wet floodplain"])]
```

```python
df_environment
```

##### 4.3.2 Lithification

```python
df_lithification = df_environment[df_environment['lithification1'] != 'unlithified']
```

```python
df_lithification
```

### 4.4 合并列

因为genus列中有重名的，即两个不同的class，genus名相同。
故重新建立一列，这个列由class和genus组成。

```python
df_lithification['class genus'] = df_lithification['class'] + df_lithification['genus']
```

```python
df_lithification
```

# 5. 数据排序与导出

### 5.1 数据排序

```python
df_final = df_lithification.sort_index()
```

```python
df_final
```
